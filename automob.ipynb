```python
import pandas as pd
import numpy as np
```

```python
auto['normalized-losses'].loc[auto['normalized-losses'] == '?'].count()
```

```python
# Set the missing value to mean of normalized losses 
# convert the datatype to integer

nl = auto['normalized-losses'].loc[auto['normalized-losses'] != '?']
nlmean = nl.astype(str).astype(int).mean()
auto['normalized-losses'] = auto['normalized-losses'].replace('?',nlmean).astype(int)
auto['normalized-losses'].head()
```

```python
auto['price'].str.isnumeric().value_counts()
```

```python

auto['price'].loc[auto['price'].str.isnumeric() == False]
```

```python
price = auto['price'].loc[auto['price'] != '?']
pmean = price.astype(str).astype(int).mean()
auto['price'] = auto['price'].replace('?',pmean).astype(int)
auto['price'].head()
```

```python
# horse power
auto['horsepower'].str.isnumeric().value_counts()
horsepower = auto['horsepower'].loc[auto['horsepower'] != '?']
hpmean = horsepower.astype(str).astype(int).mean()
auto['horsepower'] = auto['horsepower'].replace('?',pmean).astype(int)
```

```python
#outlier of horse power

auto.loc[auto['horsepower'] > 10000]
```

```python
auto[np.abs(auto.horsepower-auto.horsepower.mean())<=(3*auto.horsepower.std())]
```

```python
#bore 
auto['bore'].loc[auto['bore'] == '?']
```

```python
# replace the non numeric value to null
auto['bore'] = pd.to_numeric(auto['bore'],errors='coerce')
auto.dtypes
```

```python
# same as above
auto['stroke'] = pd.to_numeric(auto['stroke'],errors='coerce')
auto.dtypes
```

```python
auto['peak-rpm'] = pd.to_numeric(auto['peak-rpm'],errors='coerce')
auto.dtypes
```

```python
# remove the records which are having ?

auto['num-of-doors'].loc[auto['num-of-doors'] == '?']
auto = auto[auto['num-of-doors'] != '?']
auto['num-of-doors'].loc[auto['num-of-doors'] == '?']
```

```python
numericauto = auto.select_dtypes(include=['int64', 'float64'])
numericauto
```

```python
categoricalauto = auto.select_dtypes(include=['object'])
categoricalauto.head()
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='most_frequent', missing_values=np.nan)
imputer = imputer.fit(categoricalauto[['make','fuel-type','aspiration','num-of-doors','body-style','drive-wheels','engine-location','engine-type','num-of-cylinders','fuel-system']])
categoricalauto[['make','fuel-type','aspiration','num-of-doors','body-style','drive-wheels','engine-location','engine-type','num-of-cylinders','fuel-system']] = imputer.transform(categoricalauto[['make','fuel-type','aspiration','num-of-doors','body-style','drive-wheels','engine-location','engine-type','num-of-cylinders','fuel-system']])
```

```python
imputer = SimpleImputer(strategy='mean', missing_values=np.nan)
imputer = imputer.fit(numericauto[['symboling','normalized-losses','wheel-base','length','width','height','curb-weight','engine-size','bore','stroke','compression-ratio','horsepower','peak-rpm','city-mpg','highway-mpg','price']])
numericauto[['symboling','normalized-losses','wheel-base','length','width','height','curb-weight','engine-size','bore','stroke','compression-ratio','horsepower','peak-rpm','city-mpg','highway-mpg','price']] = imputer.transform(numericauto[['symboling','normalized-losses','wheel-base','length','width','height','curb-weight','engine-size','bore','stroke','compression-ratio','horsepower','peak-rpm','city-mpg','highway-mpg','price']])
```

```python
numericauto.isnull().sum()
```

```python
numeric_columns = ['symboling','normalized-losses','wheel-base','length','width','height','curb-weight','engine-size','bore','stroke','compression-ratio','horsepower','peak-rpm','city-mpg','highway-mpg','price']
numericdataframe = numericauto[numeric_columns]
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0, 1))
numeric = scaler.fit_transform(numericdataframe)
```

```python
#dataframe after scaling
Tnumericdf = pd.DataFrame(data=numeric,columns=numeric_columns)
#dataframe after scaling
Tnumericdf.head()
```

```python
categorical_columns=['make','fuel-type','aspiration','num-of-doors','body-style','drive-wheels','engine-location','engine-type','num-of-cylinders','fuel-system']
cdf = categoricalauto[categorical_columns]
dummies_df = pd.get_dummies(cdf)
dummies_df.head(20)
```

```python
#creating dummy variables for nominal categorical variable using get_dummies()
from sklearn.preprocessing import LabelEncoder
targetdf=numericauto[['price']]
le =  LabelEncoder()
targetdf['price'] = le.fit_transform(targetdf.price)
targetdf.head()
```

```python
#concat the target datafarme and categorical dataframe
df1=pd.concat([dummies_df,targetdf],axis=1)
df1
```

```python
#concat the numeric dataframe and categorical dataframe
finaldf = pd.concat([numericauto,df1],axis=1)
finaldf
```

```python
from sklearn.model_selection import train_test_split
X=finaldf.iloc[:,:-1];
y=finaldf.iloc[:,-1];
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=5)
```

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm
```

```python
lm.fit(X_train,y_train)
```

```python
lm.score(X_train,y_train)
```

```python
lm.score(X_test,y_test)
```

```python
from sklearn.metrics import mean_squared_error
y_pred = lm.predict(X_test)
print(mean_squared_error(y_test,y_pred))
```

```python
from sklearn.metrics import mean_squared_error
y_pred = lm.predict(X_test)
print(mean_squared_error(y_test,y_pred,squared=False))
```

```python
from sklearn.metrics import mean_absolute_error
y_pred = lm.predict(X_test)
print(mean_absolute_error(y_test,y_pred))
```

```python
from sklearn.metrics import r2_score
print(r2_score(y_test,y_pred))
```
